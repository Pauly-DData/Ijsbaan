#### In deze notebook zullen we de voor IJsbaan proberen te bepalen wat het goedkoopste optie was mbt contracten (variabel versus markt)

Hiervoor zullen we de volgende vragen gaan betantwoorden:

Wat is goedkoper voor de klant?

A.	<em>Afrekening via dynamische propositie, dus werkelijk verbruik per uur i.c.m. met de marktprijs</em>

B.	<em>Afrekening via gewogen gemiddelde (profiel) propositie, dus alleen het totale verbruik nemen en dat profileren over de periode (december en januari apart)</em>

we halen nu de data op uit sql met de volgende query: 
<code>
SELECT 
    FORMAT(TR.StartDate, 'dd-MM-yyyy HH:mm') as 'Start datum', 
    FORMAT(TR.EndDate, 'dd-MM-yyyy HH:mm') as 'Eind datum', 
    TR.UsageLDN as 'Verbruik in kW/h' 
FROM 
    CRM.TelemetryReading TR
INNER JOIN 
    CRM.Connection C ON C.EAN = TR.EAN
WHERE 
    TR.EAN = 'xxxxxx'
    AND TR.StartDate >= '2023-12-01' 
    AND TR.StartDate <= '2024-01-31' 
	--AND TR.UsageLDN != 0
ORDER BY
    TR.StartDate ASC; 
</code>   


```python
import pandas as pd
```

```python
df = pd.read_csv(r"telemetry.csv")
df.head()
```

```python
# we zullen eerst met pd de data naar dt omzetten.
df['Start datum'] = pd.to_datetime(df['Start datum'], format='%d-%m-%Y %H:%M')
df['Eind datum'] = pd.to_datetime(df['Eind datum'], format='%d-%m-%Y %H:%M')

# een nieuwe kolom maken voor de aangepaste tijd
df['Adjusted Hour'] = df['Start datum'].dt.floor('h')

# vervolgens berkenen we de som per uur
sum_per_hour = df.groupby('Adjusted Hour')['Verbruik in kW/h'].sum().reset_index()

# nu zetten we de aangepaste tijd weer om naar hoe we het eerst hadden
sum_per_hour['Start datum'] = sum_per_hour['Adjusted Hour']
sum_per_hour['Eind datum'] = sum_per_hour['Adjusted Hour'] + pd.Timedelta(hours=1)

# vebruik in kwh naar totaal verbruik per kwh/h
sum_per_hour.rename(columns={'Verbruik in kW/h': 'Totaal Verbruik in kW/h'}, inplace=True)

# en nu hebben we een nieuwe df waarbij we per uur het vebruik hebben ipv per kwartier
ijsbaan_uur = sum_per_hour[['Start datum', 'Eind datum', 'Totaal Verbruik in kW/h']]

#we pakken even een random rij om te bekijken
#print(ijsbaan_uur.iloc[150])

# en nu printen we de hele mikmak in uit in onze nieuwe df genaamd ijsbaan_uur
ijsbaan_uur

```

#### marktprijzen


We gaan nu de marktprijzen zoeken voor dezelfde momenten zodat we de dataset kunnnen gaan joinen om een marktprijs berekening mogelijk te maken. We zullen ook alvast de data (start & end date) naar dt verzetten zodat het mogelijk is om te joinen.

hiervoor gebruiken we het resultaat van de volgende query:
<code>
SELECT 
	FORMAT(MP.StartDate, 'dd-MM-yyyy HH:mm') as 'Start datum', 
    FORMAT(MP.EndDate, 'dd-MM-yyyy HH:mm') as 'Eind datum',
	Price / 1000 as PricePerKwH
FROM 
	CRM.MarketPrice as MP
WHERE 
	CONVERT(date, PriceDate AT TIME ZONE 'Central European Standard Time') >= '2023-12-01'
	AND CONVERT(date, PriceDate AT TIME ZONE 'Central European Standard Time') < '2024-01-31'
	AND MarketPriceType IN ('ApxDayAheadHour')
ORDER BY
	MP.StartDate ASC
</code>



```python
mp = pd.read_csv(r"marktprijs_uur.csv")

mp['Start datum'] = pd.to_datetime(df['Start datum'], format='%d-%m-%Y %H:%M')
mp['Eind datum'] = pd.to_datetime(df['Eind datum'], format='%d-%m-%Y %H:%M')

print(mp)
```

we zien dat de marktprijs per uur beschikbaar is en dat de telemetry data per kwartier is geveven daarom zullen we een sum van het vebruik doen zodat we per uur een verbruik hebben.

```python
#dan gaan we nu door naar de grote join show:
final_df = ijsbaan_uur.join(mp, how='inner', lsuffix='ijsbaan_uur', rsuffix='mp')

final_df = final_df.drop(['Start datummp', 'Eind datummp'], axis=1)

final_df['Total Cost'] = final_df['Totaal Verbruik in kW/h'] * final_df['PricePerKwH']


final_df



```

```python
totaal_verbruik = final_df['Totaal Verbruik in kW/h'].sum()
print(f"Het totale verbruik in deze periode is: {totaal_verbruik} kw/h")
```

```python
# dan nu de laatste stap; optellen van de totale kosten.
Totaal_kosten_marktprijs = final_df['Total Cost'].sum()
print(f"De totale kosten voor de periode in december en januari dat er werd vebruikt zijn: â‚¬{Totaal_kosten_marktprijs}")
```

Our above dataset looks great; now need to callculate the weighted average marketprice for dec and therefor we will be using the following query:

<code>
--Variables
DECLARE @FirstMonth int
SET @FirstMonth = 202312

DECLARE @LastMonth int
SET @LastMonth = 202312

DECLARE @MWhtom3 decimal(12,10)
SET @MWhtom3 = 0.0097694444


--Electricity

IF OBJECT_ID('tempdb..#WeightedAverageMarketPricesMonthFractionsElectricity') IS NOT NULL 
DROP TABLE #WeightedAverageMarketPricesMonthFractionsElectricity;

SELECT * INTO #WeightedAverageMarketPricesMonthFractionsElectricity
FROM (
	SELECT MonthId, Profile, SUM(CONVERT(DECIMAL(11,10),Fraction)) AS MonthlyFraction
	FROM dwh.ProfileDto
	WHERE MonthId BETWEEN @FirstMonth AND @LastMonth
	AND ProfileType = 'INT'
	AND MarketSegment = 'Electricity'
	AND FractionType = 2 --Standaard, niet dynamisch
	AND EnergyDirectionUsageType = 2 --Levering profiel voor afname zonder invoeding
	GROUP BY MonthId, Profile

) AS X



SELECT
X.ProfileDescription
,X.MonthId
,X.[WeightedAveragePrice(EUR/MWh] / 1000.0 AS [WeightedAveragePrice(EUR/kWh]
,X.[WeightedAveragePrice(EUR/MWh]
,X.NumberOfFractions
,Y.ExpectedNumberOfFractions

FROM (

	SELECT PF.ProfileDescription, PF.MonthId, SUM(PF.Fraction / (MonthlyFraction * 1.0) * MP.Price) AS [WeightedAveragePrice(EUR/MWh]
	,COUNT(PF.StartDate) AS NumberOfFractions
	FROM dwh.ProfileDto PF
	LEFT JOIN #WeightedAverageMarketPricesMonthFractionsElectricity WMF ON WMF.Profile = PF.Profile AND PF.MonthId = WMF.MonthId
	LEFT JOIN Ecedo.DWH_MarketPrice MP ON MP.MarketPriceType = 'ApxDayAheadHour' 
		AND CONVERT(date,MP.StartDate) = CONVERT(date,PF.StartDate) 
		AND DATEPART(HOUR,MP.StartDate) = DATEPART(HOUR,PF.StartDate)
	WHERE PF.MonthId BETWEEN @FirstMonth AND @LastMonth
	AND PF.ProfileType = 'INT'
	AND PF.MarketSegment = 'Electricity'
	AND FractionType = 2 --Standaard, niet dynamisch
	AND EnergyDirectionUsageType = 2 --Levering profiel voor afname zonder invoeding
	GROUP BY PF.ProfileDescription, PF.MonthId

) AS X

INNER JOIN (

	SELECT TT.MonthId, 4 * DATEDIFF(hour,CONVERT(datetime,TT.[Date]) AT TIME ZONE 'Central European Standard Time' AT TIME ZONE 'UTC',CONVERT(datetime,DATEADD(month,1,TT.[Date])) AT TIME ZONE 'Central European Standard Time' AT TIME ZONE 'UTC') AS ExpectedNumberOfFractions
	FROM DWH.TimeTable TT
	WHERE TT.[Day] = 1
) AS Y ON X.MonthId = Y.MonthId

ORDER BY X.MonthId, X.ProfileDescription



SELECT PF.MonthId, COUNT(*) AS MissingPricesElectricity
FROM dwh.ProfileDto PF
LEFT JOIN Ecedo.DWH_MarketPrice MP ON MP.MarketPriceType = 'ApxDayAheadHour' 
	AND CONVERT(date,MP.StartDate) = CONVERT(date,PF.StartDate) 
	AND DATEPART(HOUR,MP.StartDate) = DATEPART(HOUR,PF.StartDate)
WHERE PF.MonthId BETWEEN @FirstMonth AND @LastMonth
AND PF.ProfileType = 'INT'
AND PF.MarketSegment = 'Electricity'
AND MP.Price IS NULL
AND FractionType = 2 --Standaard, niet dynamisch
AND EnergyDirectionUsageType = 2 --Levering profiel voor afname zonder invoeding
GROUP BY PF.MonthId
</code>

the result of the above query for E3c is â‚¬0.081427 per kWh.

```python
fracties_dec_clean['kosten_dec_kwartier'] = fracties_dec_clean['fractie_maand_vebruik_dec'] * 0.081427

Kosten_dec_gewogen_gem = fracties_dec_clean['kosten_dec_kwartier'].sum()

print(f"De kosten voor het vebruik van Ijsbaan met profiel E3C voor de maand december in 2023 zijn: â‚¬{Kosten_dec_gewogen_gem}")
```

```python
fracties_jan_clean['kosten_jan_kwartier'] = fracties_jan_clean['fractie_maand_vebruik_jan'] * 0.079207

Kosten_jan_gewogen_gem = fracties_jan_clean['kosten_jan_kwartier'].sum()

print(f"De kosten voor het vebruik van Ijsbaan met profiel E3C voor de maand januari in 2024 zijn: â‚¬{Kosten_jan_gewogen_gem}")
```
